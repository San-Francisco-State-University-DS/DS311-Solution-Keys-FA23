## d. Create two tables to summarize 
* 1) average mpg for each cylinder class and 
* 2) the standard deviation of mpg for each gear class.

```r
# Average mpg for each cyl class -- Tidyverse Lecture Notes; 
mtcars %>% 
    group_by(cyl) %>%
    summarize(mean_mpg_cyl=mean(mpg))

# The Std.ev mpg for each gear class
mtcars %>% 
    group_by(gear) %>%
    summarize(mean_mpg_gear=mean(mpg))
```

```r
ggplot(
    data = PlantGrowth,
    aes(x=index, y=weight,  # specify the x and y axis;
    color=group,            # specify changing color based on group;
    size=weight,            # specify changing size of marker based on weight;
    shape=group)            # specify cahnging shape of marker based on group;
) + geom_point() +          # define the visualiation type is scatterplot;
    ggtitle('Group & weight') + xlab('index') + ylab('weight') # configure title and labels for x&y axis;
```

```r
# alias base layer of viualization
base <- ggplot(PlantGrowth, aes(x=weight, color=group, size=weight, shape=group,fill=group))+geom_density(alpha=0.4)
base+theme_classic()
```

```r
housingData <- read.csv('housing.csv')
housingData <- subset(housingData, 
                      select = c("Neighborhood", "Market.Value.per.SqFt", "Boro", "Year.Built"))
housingData <- na.omit(housingData)      
```

```r
head(housingData)
```

```r
# Here, we only have 2 continuous variable;
# Other 2 variables are discrete variables; 
stargazer(
    housingData,
    type='text', 
    median = TRUE
)
```

```r
housingData %>%
    group_by(Neighborhood) %>%
    summary()
```

```r
housingData %>%                             
  group_by(Boro) %>% 
  summarize(min = min(Market.Value.per.SqFt),
            q1 = quantile(Market.Value.per.SqFt, 0.25),
            median = median(Market.Value.per.SqFt),
            q3 = quantile(Market.Value.per.SqFt, 0.75), 
            max = max(Market.Value.per.SqFt))
```
